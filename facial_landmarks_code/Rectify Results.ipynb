```python
# USAGE
# python facial_landmarks.py --shape-predictor shape_predictor_68_face_landmarks.dat --image images/example_01.jpg 

# import the necessary packages
from imutils import face_utils
import os
import numpy as np
import argparse
import imutils
import dlib
import cv2
from matplotlib.pyplot import imshow
%matplotlib inline

directory = "../cartoons_2_test/images/"
images = os.listdir(directory)
images.sort()
```

```python
def get_keypoints(image_name):
    # initialize dlib's face detector (HOG-based) and then create
    # the facial landmark predictor
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

    # load the input image, resize it, and convert it to grayscale
    image = cv2.imread(image_name)
    image = imutils.resize(image, width=500)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # detect faces in the grayscale image
    rects = detector(gray, 1)
    if len(rects) == 0:
        return None
    # loop over the face detections
    for (i, rect) in enumerate(rects):
        # determine the facial landmarks for the face region, then
        # convert the facial landmark (x, y)-coordinates to a NumPy
        # array
        shape = predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)
        for (x, y) in shape:
            cv2.circle(image, (x, y), 2, (255, 0, 0), -1)
        imshow(image)
        return shape
```

```python
# Get the keypoints of all of the images
people_and_keypoints = {}

for image in images:
    image_person = image.split("-")[0]
    image_type = image.split("-")[1][:-4]
    if image_type == "targets":
        continue
    keypoints = get_keypoints(directory + image)
    if keypoints != None:
        print image
    
    if image_person not in people_and_keypoints:
        people_and_keypoints[image_person] = {}
    people_and_keypoints[image_person][image_type] = keypoints
        
```
